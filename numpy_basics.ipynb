```python
# Broadcasting is the ability of numpy to perform arithmetic operations on dissimilar array
```

```python
#Scenario 1
a1=np.arange(8).reshape(2,4)
a2=np.arange(8,16).reshape(2,4)

print(a1)
print(a2)
```

```python
a1+a2
```

```python
#Scenario 2
a3=np.arange(9).reshape(3,3)
a4=np.arange(3).reshape(1,3)

print(a3)
print(a4)
```

```python
a3+a4 #here array a4 consisting of one row will get summed with all three rows of array a3
#this is called broadcasting
```

```python
#Rules of Broadcasting
```

```python
a1=np.arange(3).reshape(1,3)
a2=np.arange(3).reshape(3,1) #same no of rows and columns so perform hoga operation

print(a1)
print(a2)
```

```python
a1+a2
```

```python
a3=np.arange(4).reshape(1,4)
a4=np.arange(12).reshape(3,4) #atleast same no of columns so perform hoga operation

print(a3)
print(a4)
```

```python
a3+a4
```

```python
a5=np.arange(3).reshape(1,3)
a6=np.arange(16).reshape(4,4) #teeno disimilar hai lekin row one toh perform nahi hoga

print(a5)
print(a6)
```

```python
# a5+a6 #error throw hoga

#abtak rule of thumb was if ek value 1 rahega and baaki 2 same rahenge toh run hoga
```

```python
# if they are of diff dimensions but number are matching then operation will be performed
a7=np.arange(4)
a8=np.arange(20).reshape(5,4)

print(a7)
print(a8)
print("\n",a7+a8)
```
